# ***Feature Engineering***

```python
# Adding a new column with the range of hematocrit for men range from 41% to 50%. Normal level for women is 36% to 48%

df['Hemato_range'] = np.where((df['SEX'] == 'M') & (df['HAEMATOCRIT'] >= 41) & (df['HAEMATOCRIT'] <= 50), 'Normal',
                               np.where((df['SEX'] == 'F') & (df['HAEMATOCRIT'] >= 36) & (df['HAEMATOCRIT'] <= 48), 'Normal',
                                       'Abnormal'))
df
```

```python
# Adding a new column with "According to the NCBI, the normal hemoglobin (Hb) level for males is 14–18 grams per deciliter (g/dl) and for females is 12–16 g/dl"

df['Heamoglobin_range'] = np.where((df['SEX'] == 'M') & (df['HAEMOGLOBINS'] >= 14) & (df['HAEMOGLOBINS'] <= 18), 'Normal',
                               np.where((df['SEX'] == 'F') & (df['HAEMOGLOBINS'] >= 12) & (df['HAEMOGLOBINS'] <= 16), 'Normal',
                                       'Abnormal'))
df
```

```python
# Adding a new column with "for women, the average platelet count is between 157,000 to 371,000 per microliter of blood. For men, the average is between 135,000 to 317,000 per microliter of blood"

df['Thrombocyte_range'] = np.where((df['SEX'] == 'M') & (df['THROMBOCYTE'] >= 135) & (df['THROMBOCYTE'] <= 317), 'Normal',
                               np.where((df['SEX'] == 'F') & (df['THROMBOCYTE'] >= 157) & (df['THROMBOCYTE'] <= 371), 'Normal',
                                       'Abnormal'))
df
```

```python
# Adding a new column with 5,000 to 19,000/mm3 for children 2 months to 6 years old. 4,800 to 10,800/mm3 for children 6 to 18 years old. 4,500 to 10,500/mm3 for adults.

df['leukocyte_range'] = np.where((df['AGE'] <= 6) & (df['LEUCOCYTE'] >= 5) & (df['LEUCOCYTE'] <= 19), 'Normal',
                               np.where((df['AGE'] <=18) & (df['LEUCOCYTE'] >= 4.8) & (df['LEUCOCYTE'] <= 10.8), 'Normal',
                                        np.where((df['AGE'] > 18) & (df['LEUCOCYTE'] >= 4.5) & (df['LEUCOCYTE'] <= 10.5), 'Normal',
                                       'Abnormal')))


df
```

```python
# Adding a new column with MCH range should be in between 25.4 to 34.6 pg/cell for all ages and genders

df['MCH_range'] = np.where((df['MCH'] >= 25.4) & (df['MCH'] <= 34.6), 'Normal',
                               'Abnormal')
df
```

```python
# Adding a new column with MCHC range should be in between 31% to 36% Hb/cell for all ages and genders

df['MCHC_range'] = np.where((df['MCHC'] >= 31) & (df['MCHC'] <= 36), 'Normal',
                               'Abnormal')
df
```

```python
# Adding a new column with MCV range should be in between 80 to 100 µm3 Hb/cell for all ages and genders

df['MCV_range'] = np.where((df['MCV'] >= 80) & (df['MCV'] <= 100), 'Normal',
                               'Abnormal')
df
```

```python
# Adding a new column with "for Male: 4.7 to 6.1 million cells per microliter (cells/mcL) Female: 4.2 to 5.4 million cells/mcL.""

df['Erythrocyte_range'] = np.where((df['SEX'] == 'M') & (df['ERYTHROCYTE'] >= 4.7) & (df['ERYTHROCYTE'] <= 6.1), 'Normal',
                               np.where((df['SEX'] == 'F') & (df['ERYTHROCYTE'] >= 4.2) & (df['ERYTHROCYTE'] <= 5.4), 'Normal',
                                       'Abnormal'))
df
```

```python
df.info()
```

Converting the categorical to numarical

```python
df['SEX'] = df['SEX'].apply(lambda x:1 if 'M' in x else 0)
df['Hemato_range'] = df['Hemato_range'].apply(lambda x:0 if 'Normal' in x else 1)
df['Heamoglobin_range'] = df['Heamoglobin_range'].apply(lambda x:0 if 'Normal' in x else 1)
df['Thrombocyte_range'] = df['Thrombocyte_range'].apply(lambda x:0 if 'Normal' in x else 1)
df['leukocyte_range'] = df['leukocyte_range'].apply(lambda x:0 if 'Normal' in x else 1)
df['MCH_range'] = df['MCH_range'].apply(lambda x:0 if 'Normal' in x else 1)
df['MCHC_range'] = df['MCHC_range'].apply(lambda x:0 if 'Normal' in x else 1)
df['MCV_range'] = df['MCV_range'].apply(lambda x:0 if 'Normal' in x else 1)
df['Erythrocyte_range'] = df['Erythrocyte_range'].apply(lambda x:0 if 'Normal' in x else 1)
```

```python
X = df.drop(columns = ['SOURCE'], axis = 1)
y = df['SOURCE']
```

# ***Splitting Data for Test and Train***

```python
# Importing Libraries and splitting data for training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

# ***Scaling the data with StandardScaler***

```python
# Importing Librarires and scaling the data using StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# **Importing a Classifiers, Matrix and Fitting the models**

```python
# Importing the Libraries
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
# Importing classifier/Regressors
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
# Fitting the model in LogisticRegression and finding the Accuracy
Logistic_Regression = LogisticRegression()
Logistic_Regression.fit(X_train_scaled, y_train)

y_pred = Logistic_Regression.predict(X_test_scaled)
cm= confusion_matrix(y_test, y_pred)
print(cm)
print("Accuracy Score: ",accuracy_score(y_test,y_pred))
```

```python
# Fitting the model in RandomForestClassifier and finding the Accuracy

Random_Forest_Classifier = RandomForestClassifier()
Random_Forest_Classifier.fit(X_train_scaled, y_train)

y_pred = Random_Forest_Classifier.predict(X_test_scaled)
cm= confusion_matrix(y_test, y_pred)
print(cm)
print("Accuracy Score: ",accuracy_score(y_test,y_pred))
```

```python
# Fitting the model in DecisionTreeClassifier and finding the Accuracy

Decision_Tree_Classifier = DecisionTreeClassifier()
Decision_Tree_Classifier.fit(X_train_scaled, y_train)

y_pred = Decision_Tree_Classifier.predict(X_test_scaled)
cm= confusion_matrix(y_test, y_pred)
print(cm)
print("Accuracy Score: ",accuracy_score(y_test,y_pred))
```

```python
# Fitting the model in SVC and finding the Accuracy
svc = SVC()
svc.fit(X_train_scaled, y_train)

y_pred = svc.predict(X_test_scaled)
cm= confusion_matrix(y_test, y_pred)
print(cm)
print("Accuracy Score: ",accuracy_score(y_test,y_pred))
```

```python
# Fitting the model in KNeighborsClassifier and finding the Accuracy
KNeighbors_Classifier = KNeighborsClassifier(n_neighbors=5)
KNeighbors_Classifier.fit(X_train_scaled, y_train)

y_pred = KNeighbors_Classifier.predict(X_test_scaled)
cm= confusion_matrix(y_test, y_pred)
print(cm)
print("Accuracy Score: ",accuracy_score(y_test,y_pred))
```

```python
# Fitting the model in GaussianNB and finding the Accuracy
Gaussian_NB = GaussianNB()
Gaussian_NB.fit(X_train_scaled, y_train)

y_pred = Gaussian_NB.predict(X_test_scaled)
cm= confusion_matrix(y_test, y_pred)
print(cm)
print("Accuracy Score: ",accuracy_score(y_test,y_pred))
```

# **Using the Neural Networks and model Fitting**

```python
# Importing models & Libraries
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
import numpy as np

# Sample data
X_train_nn = np.array(X_train_scaled)
y_train_nn = LabelEncoder().fit_transform(y_train)
X_test_nn = np.array(X_test_scaled)

# Initialize a simple Neural Network model
model = Sequential()
model.add(Dense(8, input_dim=X_train_nn.shape[1], activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train_nn, y_train_nn, epochs=10, batch_size=32, validation_split=0.2)

# Make predictions on the test set
y_pred_nn = model.predict(X_test_nn).flatten()
```
